```python
import numpy as np
import pandas as pd
```

```python
AD = pd.read_csv('AD.csv')
CN = pd.read_csv('CN.csv')
```

```python
#211, 215
AD.drop("Unnamed: 0", axis=1, inplace=True)
AD['Diagnosis'] = 1
CN.drop("Unnamed: 0", axis=1, inplace=True)
CN['Diagnosis'] = 0

data = AD.append(CN)
```

```python
meta = pd.read_csv('meta_score.csv')
meta = meta[["Subject ID", "MMSE Total Score", "Global CDR"]]
meta = meta.rename(columns={"Subject ID": "subj_id", "MMSE Total Score": "MMSE", "Global CDR": "CDR" })
```

```python
meta.drop_duplicates(subset='subj_id', inplace=True)
```

```python
data = data.merge(meta, on='subj_id')
```

```python
data.to_csv('final_AD_CN', index=False)
```
